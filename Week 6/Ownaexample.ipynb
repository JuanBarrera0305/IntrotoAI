```python
# üèÉ Track  Logic Puzzle
# Goal: Simulate training decisions and evaluate outcomes like injury risk, endurance, and peak performance

from itertools import product

# Define proposition symbols:
# I: Indoor, O: Outdoor, S: Sprint, T: Tempo, R: Rest
# N: Good Nutrition, L: Good Sleep
# E: Endurance improves, X: Injury risk, P: Peak in May

# All combinations of boolean values for 10 variables
symbols = ['I', 'O', 'S', 'T', 'R', 'N', 'L', 'E', 'X', 'P']


def rule1(O, S):  
    return not (O and S) or True  

def rule2(T, N, L):  
    return not (T and (N or L)) or True  

def rule3(R):  
    return not R or True  

def rule4(E, X):  
    return not (E and not X) or True  

def rule5(I, S, N, L):  
    return not (I and S and not N and not L) or True  

def rule8(T, N, L, E):  
    return not (T and (N or L)) or E

def rule9(O, S, I, R, N, L): 

    return (O and S) or (I and S and not N and not L) or (not R)

def rule10(E, X, P):  
    return not (E and not X) or P

results = []
for values in product([True, False], repeat=7):  
    I, O, S, T, R, N, L = values

    E = T and (N or L)  
    X = rule9(O, S, I, R, N, L)  
    P = (E and not X)  

    if rule5(I, S, N, L) and rule8(T, N, L, E) and rule10(E, X, P):
        results.append({
            'I': I, 'O': O, 'S': S, 'T': T, 'R': R,
            'N': N, 'L': L, 'E': E, 'X': X, 'P': P
        })

print(" Valid Training Scenarios (that follow all rules):\n")
for r in results:
    print(r)

print("\n Summary:")
print(f"Total valid scenarios: {len(results)}")
print(f"Scenarios where Juan peaks in May (P=True): {sum(1 for r in results if r['P'])}")
print(f"Scenarios with injury risk (X=True): {sum(1 for r in results if r['X'])}")
print(f"Scenarios with endurance gain (E=True): {sum(1 for r in results if r['E'])}")
```
